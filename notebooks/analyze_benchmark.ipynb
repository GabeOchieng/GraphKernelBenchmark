```python
import pandas as pd
```

```python
class BenchmarkEntry:
    def __init__(self, name, scores, runtime):
        self.name = name
        self.scores = scores
        self.runtime = runtime
```

```python
def load_entries(dataset_path, dataset_name):
    entries = []
    with open(dataset_path, 'r') as f:
        for entry in eval(f.readline())[dataset_name]:
            entries.append(BenchmarkEntry(entry[0], entry[1][0], entry[1][1]))
    return entries
```

```python
def build_table(entries):
    data = {
        'Algorithm': [e.name for e in entries],
        'Accuracy': [max(e.scores, key=lambda x: x[0]) for e in entries],
        'Runtime in s': [e.runtime for e in entries]
    }
    return pd.DataFrame(data=data)
    
```
